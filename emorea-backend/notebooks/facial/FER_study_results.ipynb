## Preprocessing
- Rotation correction
- Crop face
- Resize
- Normalization

⚠️ Rules of Thumb

Do NOT apply histogram equalization before alignment → it makes landmarks harder to detect.

Do NOT resize early → small faces + enhancement = blurry artifacts.

Avoid double-contrast steps (e.g., CLAHE + global hist eq) → oversaturation, loss of details.

````python
import cv2
from PIL import Image
import numpy as np
from tqdm import tqdm
tqdm.pandas()

from deepface import DeepFace
from deepface.modules.preprocessing import normalize_input
from skimage.feature import hog


# Haar cascade for eye and face detection (avoid loading multiple times)
eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_eye.xml")
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

def rotate_face(img, eye_cascade=eye_cascade):
    
    # detect eyes
    eyes = eye_cascade.detectMultiScale(img, scaleFactor=1.8)

    if len(eyes) >= 2:
        # # eye centers
        eye_1, eye_2 = eyes[0], eyes[1]
        eye_center_1 = (eye_1[0] + eye_1[2] // 2, eye_1[1] + eye_1[3] // 2)
        eye_center_2 = (eye_2[0] + eye_2[2] // 2, eye_2[1] + eye_2[3] // 2)

        # left eye first
        if eye_center_1[0] > eye_center_2[0]:
            eye_center_1, eye_center_2 = eye_center_2, eye_center_1

        # rotation angle is the arctangent of the difference in y and x coordinates
        # if the person’s head is perfectly straight, the left and right eye centers have the same y-coordinate.
        # if the head is tilted, the line between the eyes will have a slope
        dy = eye_center_2[1] - eye_center_1[1]
        dx = eye_center_2[0] - eye_center_1[0]
        angle = np.degrees(np.arctan2(dy, dx))

        # center point between eyes
        center_between_eyes = ((eye_center_1[0] + eye_center_2[0]) // 2,
                               (eye_center_1[1] + eye_center_2[1]) // 2)
        print(center_between_eyes)

        # rotate image around that center
        M = cv2.getRotationMatrix2D(center_between_eyes, angle, scale=1.0)
        rotated = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

        return rotated

    else:
        return img  # no rotation if eyes not found
    
def crop_face(img):

    # detect faces
    faces = face_cascade.detectMultiScale(img, 1.8, 3)
    if len(faces) > 0:
        x, y, w, h = faces[0]
        # crop face
        img = img[y:y+h, x:x+w]

    return img


def preprocessing_image(img, rotation_correction=False, crop=False, resize=False, normalize=False, enhance=True, target_size=(48,48)):
    """
    Preprocess the image by applying rotation correction, cropping the face, resizing, and normalizing.
    
    Parameters:
    - img: The input image to preprocess.
    - rotation_correction: Whether to apply rotation correction.
    - crop_face: Whether to crop the face from the image.
    - resize: Whether to resize the image.
    - normalize: Whether to normalize the pixel values of the image.
    
    Returns:
    - The preprocessed image.
    """
    # ensure grayscale
    # Suppose 'img' is your NumPy array (can be float or RGB)
    if img.dtype != np.uint8:
        img = (img * 255).astype(np.uint8) if img.max() <= 1.0 else img.astype(np.uint8)
    if len(img.shape) == 3 and img.shape[2] == 3:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # preprocessing
    if rotation_correction:
        img = rotate_face(img)
    if crop:
        img = crop_face(img)
    if resize:
        img = cv2.resize(img, target_size)
    if enhance:

        #img = cv2.bilateralFilter(img, d=9, sigmaColor=75, sigmaSpace=75)
        
        gamma = 1.2  # <1 = darker, >1 = brighter
        img = np.array(255*(img/255)**gamma, dtype='uint8')
        img = cv2.cvtColor(img.astype(np.uint8), cv2.COLOR_GRAY2RGB)
        #img = cv2.equalizeHist(img)
        #CLAHE
        #clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
        #img = clahe.apply(img)
    if normalize:
        img = img.astype(np.float32) / 255.0
        #img = normalize_input(img, "Facenet")
    

    return img

#df_ck_test = df_ck[df_ck['Usage']=='PrivateTest'].copy()
#df_ck_test = df_raf.copy()
#df_ck_test['label'] = df_ck_test['emo']
df_fer_test = df_fer2013[df_fer2013['Usage']=='PrivateTest']
df_ck_test = df_fer_test.copy()
df_ck_test['pixels'] = df_ck_test['pixels'].apply(lambda x: np.array([int(p) for p in x.split(" ")]).reshape(48,48))

#df_ck_test['pixels'] = df_ck_test['pixels'].apply(lambda x: np.array([int(p) for p in x.split(" ")]).reshape(48,48))
#df_ck_test['corrected'] = df_ck_test['pixels'].progress_apply(preprocessing_image)
df_ck_test['pixels'] = df_ck_test['pixels'].progress_apply(preprocessing_image)


plt.figure(figsize=(5,5))
plt.imshow(df_ck_test.iloc[2].pixels, cmap='gray')
plt.show()
````

````python

from PIL import Image
import cv2
from tqdm import tqdm
from src.recognizers import FaceEmotionRecognizer


def test_face_emotion_recognizer(img_list):
    # Initialize the recognizer
    recognizer = FaceEmotionRecognizer('opencv')
    # Load the Haar Cascade classifier for face detection
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
    predictions = []
    #img = ck_imgs[0]
    for img in tqdm(img_list):
        # Ensure data type is uint8
        gray_image = img.astype(np.uint8)

        # equalize the histogram
        #gray_image = sharpen_image(gray_image)
        gray_image = apply_clahe(gray_image)
        
        # Detect faces in the image
        faces = face_cascade.detectMultiScale(gray_image, 1.3, 1)
        if len(faces) > 0:
            x, y, w, h = faces[0]
            face_img = gray_image[y:y+h, x:x+w]
            pil_image = Image.fromarray(cv2.cvtColor(face_img, cv2.COLOR_BGR2RGB))
            predictions.append(recognizer.analyze_image(pil_image)['dominant_emotion'])
        else:
            pil_image = Image.fromarray(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
            predictions.append(recognizer.analyze_image(pil_image)['dominant_emotion'])

    return predictions


````

## Testing Your Pre-trained VGG Model on Datasets with Extra Labels

Here's a breakdown of how to handle datasets with more emotion labels than your VGG model was originally trained on:

**1. Understand Your VGG Model's Output:**

* **Number of Classes:** Determine how many emotion classes your pre-trained VGG model was trained to predict. Let's say it was trained on a dataset with 7 basic emotions (e.g., anger, disgust, fear, happiness, sadness, surprise, neutral).
* **Output Layer:** Identify the activation function of the final layer (usually softmax) and the number of output units (which corresponds to the number of classes).

**2. Understand the Target Dataset's Labels:**

* **Number of Classes:** Examine the emotion labels present in the dataset you want to test on. It might have the same 7 basic emotions, or it could have more (e.g., contempt, shame, etc.), or even a different set of labels entirely.

**3. Strategies for Handling Extra Labels:**

* **Ignore Extra Labels:** If the target dataset has labels that your VGG model doesn't predict, you can simply filter out those samples during testing. This allows you to evaluate your model on the subset of emotions it *does* recognize.
    * **Implementation:** When loading the test data, only include samples whose labels are present in the set of emotions your VGG model was trained on.
* **Map Extra Labels (if semantically related):** If some of the extra labels in the target dataset have a semantic relationship to the emotions your model predicts, you could try to map them. This is subjective and might not always be accurate. For example, "frustration" might be loosely related to "anger" or "sadness," but this kind of mapping can introduce noise.
    * **Implementation:** Create a mapping dictionary. During evaluation, if a sample has an extra label, map it to the closest label your model understands.
* **Evaluate on Common Subsets:** Identify the set of emotion labels that are common between your VGG model's training data and the target dataset. Evaluate your model's performance only on these common labels. This provides a direct comparison of how well your model generalizes to shared emotion categories.
    * **Implementation:** Filter both your VGG's predictions and the ground truth labels of the target dataset to include only the common emotions.
